```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
```

```python
tokenizer = Tokenizer(BPE())
trainer = BpeTrainer(
    special_tokens=["\n"], show_progress=True, #continuing_subword_prefix="##", end_of_word_suffix="_",
)
```

```python
tokenizer.pre_tokenizer = Whitespace()
```

```python
tokenizer.train(files=["data/tinyshakespeare.txt"], trainer=trainer)
```

```python
tokenizer.get_vocab()
```

```python
output = tokenizer.encode("Your belly's answer? What!\n")
```

```python
tokenizer.save("./data/tokenizer.json")
```

```python
pretrained = Tokenizer.from_file("./data/tokenizer.json")
```

```python
pretrained_outs = pretrained.encode("Your belly's answer? What!\n")
```
